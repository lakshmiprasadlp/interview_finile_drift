## **1. `type()`**

**Purpose:** Returns the **type** (class) of an object.

```python
print(type(10))         # <class 'int'>
print(type("hello"))    # <class 'str'>
print(type([1, 2, 3]))  # <class 'list'>
```

üí° In interviews: It‚Äôs also used for dynamic type creation when given 3 arguments.

---

## **2. `isinstance()`**

**Purpose:** Checks if an object belongs to a **class or tuple of classes**.

```python
print(isinstance(10, int))           # True
print(isinstance("hi", (int, str)))  # True
print(isinstance([1,2], list))       # True
```

üí° Safer than `type()` when working with inheritance because it checks parent classes too.

---

## **3. `id()`**

**Purpose:** Returns the **unique identity number** (memory address) of an object.

```python
x = 10
y = 10
print(id(x), id(y))  # Same ID because integers are cached
```

üí° Useful for checking if two variables point to the same object in memory.

---

## **4. `bin()`**

**Purpose:** Converts an integer to its **binary string** representation.

```python
print(bin(10))  # '0b1010'
```

üí° The `'0b'` prefix indicates binary format.

---

## **5. `pow()`**

**Purpose:** Raises a number to a power.

```python
print(pow(2, 3))       # 8 (2¬≥)
print(pow(2, 3, 5))    # 3 ((2¬≥) % 5)
```

üí° Third argument is optional and used for modulo.

---

## **6. `round()`**

**Purpose:** Rounds a number to a given number of decimal places (default = 0).

```python
print(round(3.14159, 2))  # 3.14
print(round(5.67))        # 6
```

üí° Uses ‚Äúbanker‚Äôs rounding‚Äù for .5 cases (rounds to nearest even number).

---

## **7. `abs()`**

**Purpose:** Returns the **absolute value** of a number (distance from 0).

```python
print(abs(-7))     # 7
print(abs(3.14))   # 3.14
```
