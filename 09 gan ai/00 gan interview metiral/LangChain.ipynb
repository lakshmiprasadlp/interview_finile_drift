## **1. Prompts in LangChain**

**What it is:**
The instructions you give to the model, often with **placeholders** for dynamic values.

LangChain makes prompts **more than just a string** ‚Äî they can have templates, variables, and even multiple messages.

**Example:**

```python
from langchain.prompts import PromptTemplate

template = "Write a short story about {animal} in {place}."
prompt = PromptTemplate(
    input_variables=["animal", "place"],
    template=template,
)

print(prompt.format(animal="cat", place="space"))
# "Write a short story about cat in space."
```

üîπ **Why it matters:** You can reuse the same prompt with different data.

---

## **2. Structured Output in LangChain**

**What it is:**
A way to **force** the model to return answers in a specific format (like JSON, a table, or a fixed schema).

Instead of free text, you get **machine-readable output**.

**Example:**
You ask: *"Give me weather info for Paris"*
Structured output format:

```json
{
  "city": "Paris",
  "temperature": 27,
  "condition": "Sunny"
}
```

LangChain helps you enforce this by telling the LLM **exactly what format to respond in** and checking the output.

---

## **3. Output Parsers in LangChain**

**What it is:**
A helper tool that takes the **raw text** from the model and **turns it into a Python object** (or any structured format).

Think of it as a translator between the model‚Äôs response and your code.

**Example:**

```python
from langchain.output_parsers import StructuredOutputParser, ResponseSchema

schemas = [
    ResponseSchema(name="city", description="City name"),
    ResponseSchema(name="temperature", description="Temperature in Celsius"),
    ResponseSchema(name="condition", description="Weather condition"),
]

parser = StructuredOutputParser.from_response_schemas(schemas)

# Example LLM output
llm_output = '{"city": "Paris", "temperature": 27, "condition": "Sunny"}'
parsed = parser.parse(llm_output)

print(parsed)
# {'city': 'Paris', 'temperature': 27, 'condition': 'Sunny'}
```

üîπ **Why it matters:** LLMs often return text with extra words ‚Äî parsers help you clean and extract exactly what you need.

---

üí° **In short:**

* **Prompts** = Tell the model what you want.
* **Structured Output** = Tell the model exactly how to format the answer.
* **Output Parsers** = Convert the model‚Äôs answer into a usable format in your code.

## **Chains in LangChain**

**What it is:**
A **pipeline** that connects multiple steps together ‚Äî model calls, prompts, tools, or data lookups ‚Äî to solve a bigger task.
Instead of just asking the LLM once, you **chain** multiple actions in sequence.

---

### **Why chains?**

If you just have an LLM, it‚Äôs like having a calculator ‚Äî it can answer one thing at a time.
A chain lets you say:

1. Get some data.
2. Process it.
3. Pass the result to the next step.
4. Finally, output the answer.

---

### **Example:**

**Task:** Summarize today‚Äôs weather in Paris and make it sound like a poem.

**Without chains:** You‚Äôd manually ask for weather, then copy-paste it into another request for a poem.

**With a chain:**

1. **Step 1:** Call weather API ‚Üí Get "27¬∞C, sunny"
2. **Step 2:** Pass that info to LLM ‚Üí "Write a short poem about sunny 27¬∞C weather in Paris"
3. **Step 3:** Output poem

---

### **Types of Chains in LangChain**

1. **Simple LLMChain**

   * One prompt ‚Üí One model call ‚Üí One output.

   ```python
   from langchain.chains import LLMChain
   from langchain.prompts import PromptTemplate
   from langchain.llms import OpenAI

   llm = OpenAI()
   prompt = PromptTemplate(
       input_variables=["topic"],
       template="Write a poem about {topic}."
   )

   chain = LLMChain(llm=llm, prompt=prompt)
   print(chain.run("cats"))
   ```

2. **SequentialChain**

   * Multiple steps where each output is passed to the next input.

3. **RouterChain**

   * Directs questions to different chains depending on the type of query.

4. **RetrievalQA Chain**

   * Fetches relevant documents from a database (vector store) before answering.

Alright ‚Äî **Runnables** in LangChain are actually one of the **most important ‚Äúnew style‚Äù concepts** after LangChain 0.1, so let‚Äôs break it down simply.

---

## **What are Runnables in LangChain?**

A **Runnable** is anything in LangChain that can be **run** (executed) to produce an output.
It‚Äôs like a ‚Äústandard interface‚Äù for all components ‚Äî LLMs, prompts, chains, retrievers, tools, even Python functions ‚Äî so they can connect together easily.

---

### **Why they exist**

Before, LangChain had different run styles for prompts, LLMs, chains, etc., and it was a bit messy.
Now, **everything follows the same `.invoke()` / `.stream()` / `.batch()` pattern**.

---

### **Key Runnable Methods**

1. **`.invoke(input)`** ‚Üí Run once with one input.
2. **`.batch(list_of_inputs)`** ‚Üí Run multiple inputs at once.
3. **`.stream(input)`** ‚Üí Get output as it‚Äôs generated (like streaming chat).

---

### **Example: Using Runnables**

```python
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI

# Step 1: Create components (these are runnables)
prompt = PromptTemplate.from_template("Write a haiku about {topic}")
llm = ChatOpenAI()

# Step 2: Chain them using the | operator
chain = prompt | llm

# Step 3: Run (invoke)
result = chain.invoke({"topic": "sunsets"})
print(result.content)
```

üí° Here, `prompt`, `llm`, and `chain` are all **Runnables**.

---

### **Why Runnables are Powerful**

* They make **pipelines cleaner** (`|` for chaining).
* Everything works the same way whether it‚Äôs an LLM, function, retriever, or tool.
* They unify synchronous, batch, and streaming workflows.

---

**Simple analogy:**
If LangChain is a factory:

* Before runnables ‚Üí Every machine had different buttons to start it.
* After runnables ‚Üí All machines have the same green ‚ÄúStart‚Äù button (`.invoke()`), so you can connect them easily.

Alright ‚Äî let‚Äôs go over **LangChain memory types** one by one, with **advantages and disadvantages** in plain language.

---

## **1. ConversationBufferMemory**

**What it does:**
Stores the **entire** conversation history exactly as it happened.

**‚úÖ Advantages:**

* Keeps full context ‚Äî nothing is lost.
* Great for small conversations where details matter.

**‚ùå Disadvantages:**

* Can get **very large** ‚Üí may hit LLM token limits.
* More expensive for API calls (you resend all past messages every time).

---

## **2. ConversationBufferWindowMemory**

**What it does:**
Stores **only the last N messages** (sliding window).

**‚úÖ Advantages:**

* Saves tokens (and cost).
* Keeps recent context without overwhelming the LLM.

**‚ùå Disadvantages:**

* Older info gets lost once it‚Äôs outside the window.
* Bad for long conversations where past details matter.

---

## **3. ConversationSummaryMemory**

**What it does:**
Summarizes older parts of the conversation and keeps recent messages in detail.

**‚úÖ Advantages:**

* Keeps important old info while reducing size.
* Works well for long-running conversations.

**‚ùå Disadvantages:**

* Summaries can **miss small but important details**.
* Quality depends on LLM‚Äôs summarization accuracy.

---

## **4. EntityMemory**

**What it does:**
Tracks **facts about specific entities** (people, places, items).
Example: If you say *"My dog‚Äôs name is Max"* ‚Üí It remembers "dog ‚Üí Max".

**‚úÖ Advantages:**

* Perfect for chatbots that need to remember facts about users.
* Saves memory by only storing facts, not whole conversations.

**‚ùå Disadvantages:**

* Limited to factual recall, not full conversation flow.
* Needs good entity extraction ‚Äî can misinterpret names/facts.

---

### **Quick Comparison Table**

| Memory Type                        | Best For         | Pros                                    | Cons                       |
| ---------------------------------- | ---------------- | --------------------------------------- | -------------------------- |
| **ConversationBufferMemory**       | Small chats      | Full detail, easy to use                | Grows large, costly        |
| **ConversationBufferWindowMemory** | Short-term focus | Saves cost, keeps recent info           | Loses older info           |
| **ConversationSummaryMemory**      | Long chats       | Keeps key points, scalable              | Summaries may miss details |
| **EntityMemory**                   | Fact tracking    | Stores important facts, light on tokens | Not for full conversations |

---

## **1. Document Loader**

**What it is:**
A tool that **reads your data** (from PDFs, Word docs, websites, APIs, CSVs, etc.) and loads it into LangChain as text.

**Example:**

* You have a `report.pdf`.
* Document loader extracts the text so the LLM can use it.

**LangChain examples:**

* `PyPDFLoader` (for PDFs)
* `WebBaseLoader` (for websites)
* `CSVLoader` (for CSV files)

---

## **2. Text Splitter**

**What it is:**
A tool that **breaks big text into smaller chunks** so they fit into the model‚Äôs context window.
Too much text at once ‚Üí LLM can‚Äôt handle it.

**Example:**

* A 100-page PDF is split into 500 small passages of \~500 characters each.
* Each passage keeps some overlap so meaning isn‚Äôt lost.

**LangChain examples:**

* `RecursiveCharacterTextSplitter` (smart splitting by paragraphs/sentences)
* `TokenTextSplitter` (splitting by model token limits)

---

## **3. Vector Store**

**What it is:**
A **database for embeddings** (vector representations of text/images).
You can **search by meaning**, not just exact words.

**Example:**

* ‚Äúdog‚Äù and ‚Äúpuppy‚Äù are stored close together in the vector space.
* Searching for ‚Äúdog‚Äù will also find ‚Äúpuppy‚Äù documents.

**Popular vector stores in LangChain:**

* FAISS
* Pinecone
* Weaviate
* Chroma

---

## **4. Retrievers**

**What it is:**
A tool that **fetches the most relevant chunks** from your vector store when you ask a question.

**Example:**

* You ask: ‚ÄúWhat are the side effects of drug X?‚Äù
* Retriever searches the vector store and pulls the top 3 most relevant passages from your documents.

**LangChain retrievers:**

* `vectorstore.as_retriever()`
* `MultiQueryRetriever` (asks multiple search variations)
* `SelfQueryRetriever` (uses LLM to generate search queries)

---

## **5. RAG (Retrieval-Augmented Generation)**

**What it is:**
A pattern where you **combine retrieval + generation**:

1. **Retrieve** relevant info from your documents.
2. **Feed it to the LLM** so it can answer using **real data** (not just its training).

**Example Flow:**

1. You ask: ‚ÄúSummarize the latest financial report.‚Äù
2. Retriever pulls relevant PDF chunks.
3. LLM summarizes them in plain English.

üí° This is how **ChatGPT with ‚Äúcustom knowledge‚Äù** works.

---

**Simple Analogy:**

* **Document Loader** = Brings books into your library.
* **Text Splitter** = Cuts books into readable pages.
* **Vector Store** = Organizes pages by meaning.
* **Retriever** = Finds the right pages when you ask.
* **RAG** = Reads those pages and writes you an answer.

Alright ‚Äî let‚Äôs build a **super simple RAG pipeline in LangChain** with **document loader ‚Üí text splitter ‚Üí vector store ‚Üí retriever ‚Üí RAG**.

I‚Äôll use a **local text file** here for simplicity, but you can replace it with PDFs, websites, etc.

---

```python
# Install LangChain and FAISS for vector storage
# pip install langchain langchain-community faiss-cpu openai tiktoken

from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain.chains import RetrievalQA

# 1Ô∏è‚É£ Document Loader - Load text file
loader = TextLoader("my_notes.txt")  # Replace with your file
documents = loader.load()

# 2Ô∏è‚É£ Text Splitter - Break into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(documents)

# 3Ô∏è‚É£ Vector Store - Store embeddings
embeddings = OpenAIEmbeddings()  # Needs OPENAI_API_KEY
vectorstore = FAISS.from_documents(chunks, embeddings)

# 4Ô∏è‚É£ Retriever - Search relevant chunks
retriever = vectorstore.as_retriever()

# 5Ô∏è‚É£ RAG Chain - Retrieval + LLM
llm = ChatOpenAI(model="gpt-3.5-turbo")
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff"  # "stuff" = feed all retrieved docs at once
)

# Ask a question
query = "Summarize the key points from my notes."
response = qa_chain.run(query)

print(response)
```

---

### **How it works step-by-step**

1. **Document Loader** ‚Äì Loads `my_notes.txt` into LangChain.
2. **Text Splitter** ‚Äì Breaks it into small chunks with some overlap.
3. **Vector Store (FAISS)** ‚Äì Converts chunks into embeddings and stores them.
4. **Retriever** ‚Äì Finds top-matching chunks when you ask a question.
5. **RAG** ‚Äì Sends retrieved chunks to the LLM, which writes the answer.
