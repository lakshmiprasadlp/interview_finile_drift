```python
from langchain.agents.initialize import initialize_agent# ,Tool
from langchain.agents import Tool
from langchain_openai.chat_models.base import ChatOpenAI

```

```python
from langchain_community.tools.ddg_search.tool import DuckDuckGoSearchRun
from langchain_community.tools.wikipedia.tool import WikipediaQueryRun
```

```python
search_tools=Tool(
    name="DuckDuckGo Search",
    description="Search DuckDuckGo",
    func=DuckDuckGoSearchRun().run
)
from google.colab import userdata
llm=ChatOpenAI(model_name="gpt-3.5-turbo",temperature=0,api_key=userdata.get('lp'))
tools=[search_tools]

agent=initialize_agent(tools,llm,agent="zero-shot-react-description",verbose=True)
def useDuckDuckGo(q):
  responce=agent.run(q)
  print(responce)
```

```python

useDuckDuckGo("who winn 2025 ipl cricket")
```

```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

api_wrapper = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=100)
wiki_serch_tool=Tool(
    name="Wikipedia Search",
    description="Search Wikipedia",
    func=WikipediaQueryRun(api_wrapper=api_wrapper).run
)
tools=[wiki_serch_tool]
wagent=initialize_agent(tools,llm,agent="zero-shot-react-description")

def useWikiped(q):
  responce=wagent.run(q)
  print(responce)
```

```python
useWikiped("which movie hase more rattings in karanataks")
```
