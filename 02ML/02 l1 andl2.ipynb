Let's dive into **L1** and **L2 regularization**, two critical techniques used to prevent overfitting in linear regression and other machine learning models.

---

## üî∏ What is Regularization?

**Regularization** is a technique to reduce **overfitting** by adding a penalty to the model's loss function. This discourages the model from fitting noise or overly complex patterns.

---

## üîπ 1. L1 Regularization ‚Äî **Lasso Regression**

### üîπ Concept:

L1 regularization adds the **sum of the absolute values of the coefficients** to the loss function.

### üîπ Cost Function:

$$
\text{Loss}_{\text{L1}} = \text{MSE} + \lambda \sum_{j=1}^{n} |\beta_j|
$$

* $\lambda$: Regularization parameter (controls penalty strength)
* $\beta_j$: Coefficients of the model

### üîπ Key Property:

* L1 can **shrink some coefficients to exactly 0**, performing **feature selection**.
* Sparse models: Useful when only some features are important.

---

## üîπ 2. L2 Regularization ‚Äî **Ridge Regression**

### üîπ Concept:

L2 regularization adds the **sum of the squared values of the coefficients** to the loss function.

### üîπ Cost Function:

$$
\text{Loss}_{\text{L2}} = \text{MSE} + \lambda \sum_{j=1}^{n} \beta_j^2
$$

* Penalizes large weights but doesn‚Äôt shrink them to zero.

### üîπ Key Property:

* L2 **does not perform feature selection**.
* All features are kept but weights are reduced.
* Works well when all features are relevant.

---

## üî∏ Comparison Table

| Feature           | L1 (Lasso)                    | L2 (Ridge)                             |   |                  |
| ----------------- | ----------------------------- | -------------------------------------- | - | ---------------- |
| Penalty term      | ( \sum                        | \beta\_j                               | ) | $\sum \beta_j^2$ |
| Feature selection | ‚úÖ Yes                         | ‚ùå No                                   |   |                  |
| Coefficients      | Can become **zero**           | Tend to become **small**               |   |                  |
| Model simplicity  | Encourages sparsity           | Encourages small weights               |   |                  |
| Best use case     | High-dimensional data, sparse | Multicollinearity, all features useful |   |                  |

---

## üîπ 3. Elastic Net = L1 + L2

Elastic Net combines L1 and L2:

$$
\text{Loss}_{\text{ElasticNet}} = \text{MSE} + \lambda_1 \sum |\beta_j| + \lambda_2 \sum \beta_j^2
$$

Use when:

* You want feature selection (from L1)
* And want model stability (from L2)

---

## ‚úÖ Python Code Example (L1 & L2)

```python
from sklearn.linear_model import Lasso, Ridge
import numpy as np

# Simple dummy data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.2, 1.9, 3.2, 3.8, 5.1])

# L1 - Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
print("Lasso Coefficients:", lasso.coef_)

# L2 - Ridge
ridge = Ridge(alpha=0.1)
ridge.fit(X, y)
print("Ridge Coefficients:", ridge.coef_)
```
