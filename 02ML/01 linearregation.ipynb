Let's go through each concept in-depth:

---

## 🔹 1. Linear Regression (Simple Linear Regression)

### **Definition:**

Linear Regression is a statistical technique that models the relationship between:

* A **dependent variable** `y`
* A **single independent variable** `x`

It assumes a **linear relationship**:

$$
y = \beta_0 + \beta_1x + \epsilon
$$

* $\beta_0$: Intercept (value of y when x=0)
* $\beta_1$: Slope (change in y for one unit change in x)
* $\epsilon$: Error term

### **Goal:**

To find the best-fitting straight line (using **least squares**) that minimizes the difference between actual and predicted values.

### **Example Use Case:**

Predicting salary based on years of experience.

---

## 🔹 2. Multiple Linear Regression

### **Definition:**

Extends simple linear regression to use **multiple independent variables**:

$$
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n + \epsilon
$$

### **Example:**

Predict house price based on:

* Size
* Number of bedrooms
* Age of house
* Location score, etc.

### **Assumptions:**

* Linear relationship
* No multicollinearity
* Homoscedasticity (constant variance of errors)
* Normal distribution of errors
* Independence of errors

---

## 🔹 3. Polynomial Regression

### **Definition:**

Used when the relationship between the independent and dependent variable is **non-linear**.

Still linear in coefficients:

$$
y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + ... + \beta_nx^n + \epsilon
$$

### **How it's used:**

Convert features to polynomial form and apply linear regression.

### **Example:**

Growth rate of a tree over time, temperature trends, etc.

---

## 🔹 4. Performance Metrics

These evaluate how well your regression model fits the data.

### **a. Mean Absolute Error (MAE)**

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

* Average of absolute differences between actual and predicted values.
* **Interpretation:** Lower is better.
* **Advantage:** Not sensitive to outliers.

---

### **b. Root Mean Squared Error (RMSE)**

$$
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 }
$$

* Penalizes larger errors more than MAE.
* **Advantage:** Sensitive to large errors, more punishing.
* **Use case:** When large errors are unacceptable.

---

### **c. R² Score (Coefficient of Determination)**

$$
R^2 = 1 - \frac{\text{SS}_{\text{res}}}{\text{SS}_{\text{tot}}}
$$

Where:

* $\text{SS}_{\text{res}} = \sum (y_i - \hat{y}_i)^2$ → Residual sum of squares
* $\text{SS}_{\text{tot}} = \sum (y_i - \bar{y})^2$ → Total sum of squares

**Interpretation:**

* R² = 1 → Perfect prediction
* R² = 0 → No predictive power
* Can be negative if model performs worse than mean

---

### **d. Adjusted R² Score**

Used in **Multiple Linear Regression** to penalize overfitting by adding too many predictors.

$$
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)
$$

Where:

* $n$ = Number of observations
* $p$ = Number of predictors

**Key Points:**

* Adjusted R² increases only when the new variable improves model performance.
* Use it to compare models with different numbers of predictors.

---

## 🔹 Summary Table

| Metric          | Best For                | Sensitive to Outliers | Range                    | Higher/Lower is Better |
| --------------- | ----------------------- | --------------------- | ------------------------ | ---------------------- |
| **MAE**         | General fit             | No                    | $\geq 0$                 | Lower                  |
| **RMSE**        | Penalizing large errors | Yes                   | $\geq 0$                 | Lower                  |
| **R²**          | Overall model fit       | Yes                   | 0 to 1 (can be negative) | Higher                 |
| **Adjusted R²** | Multiple regression     | Yes                   | 0 to 1                   | Higher                 |

## 🔹 Python Code Example

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Sample actual vs predicted
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

# MAE
mae = mean_absolute_error(y_true, y_pred)

# RMSE
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# R² Score
r2 = r2_score(y_true, y_pred)

# Adjusted R² (manual calculation)
n = len(y_true)  # samples
p = 1            # number of features used (change accordingly)
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}")
print(f"Adjusted R²: {adjusted_r2}")
```

Here’s a **simple implementation of Linear Regression, Multiple Linear Regression, and Polynomial Regression** with performance metrics (MAE, RMSE, R², Adjusted R²) using **NumPy and scikit-learn** on a **toy dataset**.

---

## ✅ Step-by-step Simple Python Implementation

### ✅ Step 1: Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

---

### ✅ Step 2: Generate Toy Data

#### Simple Linear (1 feature)

```python
# Data for Simple Linear Regression
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.5, 3.8, 6.2, 7.6, 10.2])
```

---

### ✅ Step 3: Linear Regression

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("Linear Regression:")
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

---

### ✅ Step 4: Evaluate Metrics

```python
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
n = len(y)
p = 1  # only 1 feature
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE: {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R²: {r2:.3f}")
print(f"Adjusted R²: {adjusted_r2:.3f}")
```

---

### ✅ Step 5: Plot Linear Fit

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

---

## 🔁 Multiple Linear Regression

```python
# Multiple Linear Regression Data
X_multi = np.array([
    [1, 2],
    [2, 1],
    [3, 4],
    [4, 3],
    [5, 5]
])
y_multi = np.array([2.5, 3.1, 6.9, 7.2, 10.8])

model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)
y_multi_pred = model_multi.predict(X_multi)

print("\nMultiple Linear Regression:")
print("Coefficients:", model_multi.coef_)
print("Intercept:", model_multi.intercept_)

# Metrics
mae_m = mean_absolute_error(y_multi, y_multi_pred)
rmse_m = np.sqrt(mean_squared_error(y_multi, y_multi_pred))
r2_m = r2_score(y_multi, y_multi_pred)
n_m = len(y_multi)
p_m = X_multi.shape[1]
adjusted_r2_m = 1 - (1 - r2_m) * (n_m - 1) / (n_m - p_m - 1)

print(f"MAE: {mae_m:.3f}")
print(f"RMSE: {rmse_m:.3f}")
print(f"R²: {r2_m:.3f}")
print(f"Adjusted R²: {adjusted_r2_m:.3f}")
```

---

## 🔁 Polynomial Regression (degree 2)

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model_poly = LinearRegression()
model_poly.fit(X_poly, y)
y_poly_pred = model_poly.predict(X_poly)

print("\nPolynomial Regression (degree=2):")
print("Coefficients:", model_poly.coef_)
print("Intercept:", model_poly.intercept_)

# Metrics
mae_p = mean_absolute_error(y, y_poly_pred)
rmse_p = np.sqrt(mean_squared_error(y, y_poly_pred))
r2_p = r2_score(y, y_poly_pred)
p_p = X_poly.shape[1] - 1
adjusted_r2_p = 1 - (1 - r2_p) * (n - 1) / (n - p_p - 1)

print(f"MAE: {mae_p:.3f}")
print(f"RMSE: {rmse_p:.3f}")
print(f"R²: {r2_p:.3f}")
print(f"Adjusted R²: {adjusted_r2_p:.3f}")

# Plotting
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_poly = model_poly.predict(X_range_poly)

plt.scatter(X, y, color='green', label='Actual')
plt.plot(X_range, y_range_poly, color='orange', label='Polynomial Fit')
plt.title("Polynomial Regression (degree=2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
