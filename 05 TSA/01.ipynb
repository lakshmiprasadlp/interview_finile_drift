```python
import numpy as np
import matplotlib.pyplot as plt

# Time
time = np.arange(0, 300, 1)

# Components
trend = 0.05 * time                               # upward linear trend
seasonality = 10 * np.sin(2 * np.pi * time / 30)  # regular repeating seasonality
cycle = 7 * np.sin(2 * np.pi * time / 100)        # long-term cyclic pattern
stationary = np.random.normal(0, 1, len(time))    # stationary random process
noise = np.random.normal(0, 3, len(time))         # pure noise

# Plot all in one figure
fig, axs = plt.subplots(5, 1, figsize=(14, 14), sharex=True)

axs[0].plot(time, trend, color='blue')
axs[0].set_title("Trend (long-term increase/decrease)")

axs[1].plot(time, seasonality, color='green')
axs[1].set_title("Seasonality (repeating pattern)")

axs[2].plot(time, cycle, color='orange')
axs[2].set_title("Cyclic (long-term oscillations, not fixed period)")

axs[3].plot(time, stationary, color='purple')
axs[3].set_title("Stationary (constant mean & variance over time)")

axs[4].plot(time, noise, color='red')
axs[4].set_title("Noise (random errors, no structure)")

plt.tight_layout()
plt.show()
```
