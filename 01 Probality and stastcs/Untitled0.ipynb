## ✅ **1. Hypothesis Testing – Concept & Mechanism**

**Definition:**
Hypothesis testing is a statistical method to **make decisions or inferences about population parameters** using sample data.

---

### ✅ **Steps in Hypothesis Testing**

1. **State Hypotheses:**

   * **Null Hypothesis (H₀):** No effect, no difference (status quo).
   * **Alternative Hypothesis (H₁):** There is an effect or difference.
   * Example:
     H₀: μ = 50
     H₁: μ ≠ 50

2. **Choose Significance Level (α):**

   * Common values: 0.05 (5%), 0.01 (1%).

3. **Select Appropriate Test & Compute Test Statistic:**

   * Z-test, T-test, Chi-square, ANOVA depending on scenario.

4. **Compute p-value OR Compare Test Statistic with Critical Value:**

   * If **p ≤ α**, reject H₀.

5. **Make Decision:**

   * Reject or fail to reject H₀.

---

### ✅ **Types of Errors**

* **Type I Error (α):** Rejecting H₀ when it’s true.
* **Type II Error (β):** Failing to reject H₀ when it’s false.

## ✅ **2. Z-Test**

**When to Use:**

* Population standard deviation (σ) **known**.
* Sample size **n ≥ 30** (large sample).
* Data approximately **normally distributed**.

**Formula for One-Sample Z-Test:**

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

**Applications:**

* Test sample mean vs population mean.
* Proportion tests.

---

---

## ✅ **3. T-Test**

**When to Use:**

* Population standard deviation (σ) **unknown**.
* Sample size **n < 30** (small sample).
* Data approximately **normally distributed**.

**Formula for One-Sample T-Test:**

$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
$$

where **s** = sample standard deviation.

**Types of T-Tests:**

* **One-sample T-test:** Compare sample mean to population mean.
* **Two-sample T-test (Independent):** Compare two different group means.
* **Paired T-test:** Compare before vs after (dependent samples).

---

---

## ✅ **4. Z-Test vs T-Test (Comparison Table)**

| Feature           | Z-Test              | T-Test                         |
| ----------------- | ------------------- | ------------------------------ |
| σ (Population SD) | Known               | Unknown                        |
| Sample Size       | Large (n ≥ 30)      | Small (n < 30)                 |
| Distribution Used | Standard Normal (Z) | Student’s t-distribution       |
| Shape             | Fixed               | Varies with degrees of freedom |

---

---

### ✅ **Why Degrees of Freedom in T-Test?**

* T-distribution changes shape depending on **degrees of freedom (n-1)**.
* For large n, **T → Z distribution**.

## ✅ **3. T-Test**

**When?**

* Population **σ unknown**
* **Small sample (n < 30)**
* Normal distribution assumption

**Formula:**

$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
$$

**Types:**

* **One-Sample T-Test**: Compare sample mean vs population mean
* **Independent Two-Sample T-Test**: Compare two different groups
* **Paired T-Test**: Compare before-after in same group

---

## ✅ **4. Chi-Square Test (χ²)**

**Purpose:**

* Test **association between categorical variables**
* Test **goodness of fit** of observed data with expected data

**When to Use:**

* Data is **categorical**
* Large sample size

**Formula:**

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where O = Observed frequency, E = Expected frequency

**Applications:**

* **Chi-Square Goodness-of-Fit**: Does sample data match a distribution?
* **Chi-Square Test of Independence**: Are two variables independent?
